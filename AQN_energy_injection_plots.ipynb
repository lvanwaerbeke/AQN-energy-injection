```python
import numpy as np
from astropy.cosmology import LambdaCDM

from astropy import constants as cst
from astropy import units as u
import matplotlib.pyplot as plt
import scipy.integrate as sc

from IPython.display import display, Image

light_green = (144/255, 238/255, 144/255)

%pylab inline
```

# TT, TE and EE
### AQN Calculations from Fereshteh Majidi with CLASS

```python
data_noinj = np.loadtxt('No_Inj_07_cl.dat')
data_aqn_1g = np.loadtxt('AQN_1g_20_cl.dat')
T_cmb_0 = 2.725*u.K
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 4))

axs[0].plot(data_noinj[:,0],data_noinj[:,1]/1e-12*T_cmb_0**2*u.microKelvin**2)
axs[0].set_yscale('log')
axs[0].set_xlabel('$\ell$')
axs[0].set_ylabel('$\ell(\ell+1)/2\pi C_\ell ~~[\mu K^2]$')
axs[0].set_title('$TT$')

axs[1].plot(data_noinj[:,0],data_noinj[:,3]/1e-12*T_cmb_0**2*u.microKelvin**2)
#axs[1].set_yscale('log')
axs[1].set_xlabel('$\ell$')
#axs[1].set_ylabel('$\ell(\ell+1)/2\pi C_\ell ~~[\mu K^2]$')
axs[1].set_title('$TE$')

axs[2].plot(data_noinj[:,0],data_noinj[:,2]/1e-12*T_cmb_0**2*u.microKelvin**2)
#axs[2].set_yscale('log')
axs[2].set_xlabel('$\ell$')
#axs[2].set_ylabel('$\ell(\ell+1)/2\pi C_\ell ~~[\mu K^2]$')
axs[2].set_title('$EE$')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 4))

axs[0].plot(data_noinj[:,0],(data_noinj[:,1]-data_aqn_1g[:,1])/1e-12*T_cmb_0**2*u.microKelvin**2)
#axs[0].set_yscale('log')
axs[0].set_xlabel('$\ell$')
axs[0].set_ylabel('$\ell(\ell+1)/2\pi C_\ell ~~[\mu K^2]$')
axs[0].set_title('$TT-TT_{AQN}$')

axs[1].plot(data_noinj[:,0],(data_noinj[:,3]-data_aqn_1g[:,3])/1e-12*T_cmb_0**2*u.microKelvin**2)
#axs[1].set_yscale('log')
axs[1].set_xlabel('$\ell$')
#axs[1].set_ylabel('$\ell(\ell+1)/2\pi C_\ell ~~[\mu K^2]$')
axs[1].set_title('$TE-TE_{AQN}$')

axs[2].plot(data_noinj[:,0],(data_noinj[:,2]-data_aqn_1g[:,2])/1e-12*T_cmb_0**2*u.microKelvin**2)
#axs[2].set_yscale('log')
axs[2].set_xlabel('$\ell$')
#axs[2].set_ylabel('$\ell(\ell+1)/2\pi C_\ell ~~[\mu K^2]$')
axs[2].set_title('$EE-EE_{AQN}$')
```
